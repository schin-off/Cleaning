```python
import pandas as pd
import matplotlib.pyplot as plt

rd = pd.read_csv('cleaned_data.csv')# leio o arquivo já limpo

fig, ax = plt.subplots()#crio o grafico
ax.fill_between(rd['Date'], rd['Calories'], alpha=0.5)#defino X e Y

ax.set_title('Gráfico de Calorias')#o nome do grafico
ax.set_xlabel('Data')#o nome de X
ax.set_ylabel('Calorias')#o nome de Y
ax.grid(True)#para ter a grade
plt.xticks(rotation=45)#pra rotacionar o X pra não sobre por um ao outro

plt.show()#grafico bonitinho
rd.plot( 'Date' , 'Calories' ).grid(True)#o antigo para ter uma base
```
